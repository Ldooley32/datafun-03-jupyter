##  Initial Data Distribution for Numerical Columns

```python
#Inspect histogram by numerical column
df['petal_length'].hist()
plt.title('Histogram of Petal Length')
plt.xlabel('Petal Length')
plt.ylabel('Frequency')

# Inspect histograms for all numerical columns
df.hist()

# show all plots
plt.show()
```

### Histogram Observations 

1. **Petal Length**:
   - The histogram for petal length appears to be bimodal. The first mode is around 1.5 cm and the second is around 5 cm. The peaks correspond to different species. the range of petal lenghts span from 1 cm to 7 cm. The first mode is more concentrated while the second is more spread out, suggesting there is more veriation within the second species. 
2. **Petal Width**:
   - The histogram of petal width shows a multimodal distribution. The 4 modes peacking at .2 cm, 1.4 cm, 2cm, and 2.5cm, which indicating less veriation of petal widths than petal lengths.
3. **Sepal Length**:
   - The histogram of the sepal length is a multimodal distubution. The 3 modes peak at 5 cm, 5.5cm, 6.2 cm. The spread is from 1 cm to 8 cm showing the variations the sepal length can have. 
4. **Sepal Width**:
   - The histogram of the sepal width is unimodial. The mode peaks at 3 cm. and the spead appears to be concetrated from 2cm to 4 cm, which leaves very little variation with sepal width.

## Initial Data Transformation and Feature Engineering

```python

def rename_columns(df):
    """Rename columns of the DataFrame."""
    new_column_names = {
        'sepal_length': 'Sepal Length (cm)',
        'sepal_width': 'Sepal Width (cm)',
        'petal_length': 'Petal Length (cm)',
        'petal_width': 'Petal Width (cm)',
        'species': 'Species'
    }
    df.rename(columns=new_column_names, inplace=True)

def add_new_columns(df):
    """Add new columns 'Sepal Area' and 'Petal Area'."""
    df['Sepal Area'] = df['Sepal Length (cm)'] * df['Sepal Width (cm)']
    df['Petal Area'] = df['Petal Length (cm)'] * df['Petal Width (cm)']


# Assume df is already defined and contains the original data
# Rename columns
rename_columns(df)

# Adding new columns
add_new_columns(df)
```

## Initial Visualizations

```python

def plot_pairplot(df):
    """Plot pairplot with the new named columns."""
    sns.pairplot(df, hue="Species")
    plt.show()



# Plot pairplot
plot_pairplot(df)
```
